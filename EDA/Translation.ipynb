```python
from pororo import Pororo
```

```python
mt = Pororo(task="translation", lang="multi")
```

```python
cur_dir = '/content/drive/MyDrive/KLUE_TC/data/open/'
```

```python
import pandas as pd
```

```python
dacon_train_data = pd.read_csv(cur_dir + 'train_data_translated.csv')
dacon_topic_data = pd.read_csv(cur_dir + 'topic_dict.csv')
dacon_test_data = pd.read_csv(cur_dir + 'test_data_translated.csv')
```

```python
texts = list(dacon_test_data['title_en'])
```

```python
from tqdm import tqdm
```

```python
title_kr = []

for t in tqdm(texts):
    title_kr.append(mt(t, src="en", tgt="ko"))
```

```python
title_kr
```

```python
len(title_kr)
```

```python
def writeDict(file_name, dictionary):
    with open(file_name, 'w', encoding='utf-8') as file:
        for data in dictionary:
            file.write(data + '\n')
            
writeDict(cur_dir + "test_title_kr.txt", title_kr)
```
