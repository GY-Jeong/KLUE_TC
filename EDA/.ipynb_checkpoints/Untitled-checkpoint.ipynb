```python
dacon_title = dacon_train_data['title'].unique()
dacon_test_title = dacon_test_data['title'].unique()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoConfig
```

```python
# tokenizer = AutoTokenizer.from_pretrained('monologg/koelectra-base-v3-discriminator')
xlm_roberta_tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-large')


model_name_or_path = 'monologg/koelectra-base-v3-discriminator'
# Load pretrained model and tokenizer
config = AutoConfig.from_pretrained(
    model_name_or_path,
)

tokenizer = AutoTokenizer.from_pretrained(
    model_name_or_path,
    use_fast=True,
)

config.num_labels = 6
model = AutoModelForSequenceClassification.from_pretrained(
    model_name_or_path,
    from_tf=bool(".ckpt" in model_name_or_path),
    config=config,
)
```

```python
klue_title_tokenized = [tokenizer.encode(title) for title in klue_title]
```

```python
dacon_title_tokenized = [tokenizer.encode(title) for title in dacon_title]
dacon_title_tokenized_roberta = [xlm_roberta_tokenizer.encode(title) for title in dacon_title]
dacon_test_title_tokenized = [tokenizer.encode(title) for title in dacon_test_title]
dacon_test_title_tokenized_roberta = [xlm_roberta_tokenizer.encode(title) for title in dacon_test_title]
```

```python
dacon_title_tokenized_len = [len(element) for element in dacon_title_tokenized]
dacon_title_tokenized_len_roberta = [len(element) for element in dacon_title_tokenized_roberta]
dacon_test_title_tokenized_len = [len(element) for element in dacon_test_title_tokenized]
dacon_test_title_tokenized_len_roberta = [len(element) for element in dacon_test_title_tokenized_roberta]
```

```python
s = pd.DataFrame([dacon_title_tokenized_len, dacon_title_tokenized_len_roberta, dacon_test_title_tokenized_len, dacon_test_title_tokenized_len_roberta]).T
s.columns = ['tokenized', 'tokenized_roberta', 'test_tokenized', 'test_tokenized_roberta']
```
